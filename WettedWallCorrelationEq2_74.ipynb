```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default='plotly_dark'
import numpy as np
from scipy.optimize import minimize
```

```python
Re = np.random.uniform(2e3, 35e3, (100,))
Sc = np.random.uniform(0.6, 2.5, (100,))
Sh=0.023*Re**0.83 * Sc**0.44     # Gilliland/Sherwood correlation
Sh = np.random.normal(loc=Sh, scale=0.05*Sh)  # add "experimental error"
```

```python
# ln Sh = ln C + A ln Re + B ln Sc

X = np.column_stack([np.ones_like(Sh), np.log(Re), np.log(Sc)])
b = np.linalg.inv(X.T @ X) @ X.T @ np.log(Sh)
```

```python
def sumsq(p):
    # return np.sum((Sh - p[0]*Re**p[1] * Sc**p[2])**2)  # Constant absolute error
    return np.sum((Sh - p[0]*Re**p[1] * Sc**p[2])**2 / Sh**2)  # Constant relative error

minimize(sumsq,(0.023, 0.83, 0.44))
```

```python
print(b, np.exp(b[0]))
```

```python
fig = make_subplots()
fig.add_trace(go.Scatter3d(x=Re, y=Sc, z=Sh, mode='markers', marker_size=5, marker_color=Sh, marker_opacity=0.9, marker_line_width=1, marker_line_color='white'))

Sc_grid, Re_grid  = np.meshgrid(np.linspace(0.6, 2.5, 30), np.linspace(2e3, 35e3, 30))
Sh_grid = 0.023*Re_grid**0.83 * Sc_grid**0.44

for i, j, k in zip(Re_grid, Sc_grid, Sh_grid):
    fig.add_trace(go.Scatter3d(x=i, y=j, z=k, mode='lines', line=dict(color='#0066FF', width=2)))

Re_grid, Sc_grid  = np.meshgrid( np.linspace(2e3, 35e3, 30),np.linspace(0.6, 2.5, 30),)
Sh_grid = 0.023*Re_grid**0.83 * Sc_grid**0.44

for i, j, k in zip(Re_grid, Sc_grid, Sh_grid):
    fig.add_trace(go.Scatter3d(x=i, y=j, z=k, mode='lines', line=dict(color='#0066FF', width=2)))

fig.update_layout(width=800,height=800, showlegend=False, scene=dict(xaxis_title="Re", yaxis_title='Sc', zaxis_title="Sh"))
```
